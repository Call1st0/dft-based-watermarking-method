### Comparing 3 different GCR Methods for masking Image Wateramark

##### Method 1 - most aggressive
##### Method 2 - compromise
##### Method 3 - most watermark-safe

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# Importing CSV files - different seeds
Marked_Optimal = pd.read_csv('metricDataFrameMarked1024.csv')

Method1_Optimal = pd.read_csv('metricDataFrameGCR-1-gcr.csv')
Method2_Optimal = pd.read_csv('metricDataFrameGCR-2-gcr.csv')
Method3_Optimal = pd.read_csv('metricDataFrameGCR-3-gcr.csv')
```

```python
#Getting mean values from DataFrames; Dropping first row
Marked_OptimalMean = Marked_Optimal.mean()
Marked_OptimalMean = Marked_OptimalMean.drop(['Unnamed: 0'])

Method1_OptimalMean = Method1_Optimal.mean()
Method1_OptimalMean = Method1_OptimalMean.drop(['Unnamed: 0'])

Method2_OptimalMean = Method2_Optimal.mean()
Method2_OptimalMean = Method2_OptimalMean.drop(['Unnamed: 0'])

Method3_OptimalMean = Method3_Optimal.mean()
Method3_OptimalMean = Method3_OptimalMean.drop(['Unnamed: 0'])
```

```python
CompareMethods = pd.concat([Marked_OptimalMean, Method1_OptimalMean, Method2_OptimalMean, Method3_OptimalMean], 
                        axis = 1)
CompareMethods.columns = ['Marked', 'Method 1', 'Method 2', 'Method 3']
```

```python
CompareMethods
```

```python
method2 = pd.concat([Marked_OptimalMean, Method2_OptimalMean], 
                        axis = 1)
method2.columns = ['Marked', 'GCR']
```

```python
method2.round(2)
```

```python
CompareGCR = pd.concat([Marked_Optimal['GCR / Marked'], 
                              Method1_Optimal['GCR / Marked'], 
                              Method2_Optimal['GCR / Marked'], 
                              Method3_Optimal['GCR / Marked']], axis = 1)
CompareGCR.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareCropped = pd.concat([Marked_Optimal['Cropped'], 
                              Method1_Optimal['Cropped'], 
                              Method2_Optimal['Cropped'], 
                              Method3_Optimal['Cropped']], axis = 1)
CompareCropped.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareScaled = pd.concat([Marked_Optimal['Scaled'], 
                              Method1_Optimal['Scaled'], 
                              Method2_Optimal['Scaled'], 
                              Method3_Optimal['Scaled']], axis = 1)
CompareScaled.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareRotated = pd.concat([Marked_Optimal['Rotated'], 
                              Method1_Optimal['Rotated'], 
                              Method2_Optimal['Rotated'], 
                              Method3_Optimal['Rotated']], axis = 1)
CompareRotated.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareAffined = pd.concat([Marked_Optimal['Affined'], 
                              Method1_Optimal['Affined'], 
                              Method2_Optimal['Affined'], 
                              Method3_Optimal['Affined']], axis = 1)
CompareAffined.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareConvoluted = pd.concat([Marked_Optimal['2D Convoluted'], 
                              Method1_Optimal['2D Convoluted'], 
                              Method2_Optimal['2D Convoluted'], 
                              Method3_Optimal['2D Convoluted']], axis = 1)
CompareConvoluted.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareBlurred = pd.concat([Marked_Optimal['Blurred'], 
                              Method1_Optimal['Blurred'], 
                              Method2_Optimal['Blurred'], 
                              Method3_Optimal['Blurred']], axis = 1)
CompareBlurred.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']

CompareNoised = pd.concat([Marked_Optimal['Noised'], 
                              Method1_Optimal['Noised'], 
                              Method2_Optimal['Noised'], 
                              Method3_Optimal['Noised']], axis = 1)
CompareNoised.columns = ['Marked', 'GCR Method 1', 'GCR Method 2', 'GCR Method 3']
```
