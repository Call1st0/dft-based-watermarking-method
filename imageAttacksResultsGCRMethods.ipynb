```python
import pandas as pd
```

```python
# Importing CSV files - different seeds
Marked_10000 = pd.read_csv('TestSetGCRMethod/metricDataFrameMarked-IF10000.csv')
Marked_Optimal = pd.read_csv('TestSetGCRMethod/metricDataFrameMarked-IFOptimal.csv')

Method1_10000 = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IF10000-1-gcr.csv')
Method2_10000 = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IF10000-2-gcr.csv')
Method3_10000 = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IF10000-3-gcr.csv')

Method1_Optimal = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IFOptimal-1-gcr.csv')
Method2_Optimal = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IFOptimal-2-gcr.csv')
Method3_Optimal = pd.read_csv('TestSetGCRMethod/metricDataFrameGCR-IFOptimal-3-gcr.csv')
```

```python
#Getting mean values from DataFrames; Dropping first row
Method1_10000Mean = Method1_10000.mean()
Method1_10000Mean = Method1_10000Mean.drop(['Unnamed: 0'])

Method2_10000Mean = Method2_10000.mean()
Method2_10000Mean = Method2_10000Mean.drop(['Unnamed: 0'])

Method3_10000Mean = Method3_10000.mean()
Method3_10000Mean = Method3_10000Mean.drop(['Unnamed: 0'])

Method1_OptimalMean = Method1_Optimal.mean()
Method1_OptimalMean = Method1_OptimalMean.drop(['Unnamed: 0'])

Method2_OptimalMean = Method2_Optimal.mean()
Method2_OptimalMean = Method2_OptimalMean.drop(['Unnamed: 0'])

Method3_OptimalMean = Method3_Optimal.mean()
Method3_OptimalMean = Method3_OptimalMean.drop(['Unnamed: 0'])
```

```python
# Creating DataFrame for comparing GCR Methods - Impact Factor = 10000
dfCompare10000 = pd.concat([Marked_10000['GCR / Marked'], 
                            Method1_10000['GCR / Marked'], 
                            Method2_10000['GCR / Marked'], 
                            Method3_10000['GCR / Marked']], axis = 1)
dfCompare10000.columns = ['Marked (10000)', 'Method 1', 'Method 2', 'Method 3']
dfCompare10000
```

```python
# Creating DataFrame for comparing GCR Methods - Impact Factor = Optimal
dfCompareOptimal = pd.concat([Marked_Optimal['GCR / Marked'], 
                              Method1_Optimal['GCR / Marked'], 
                              Method2_Optimal['GCR / Marked'], 
                              Method3_Optimal['GCR / Marked']], axis = 1)
dfCompareOptimal.columns = ['Marked (Optimal)', 'Method 1', 'Method 2', 'Method 3']
dfCompareOptimal
```
