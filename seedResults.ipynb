```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# Importing CSV files - different seeds
dfSeed = pd.read_csv('seedDataFrame.csv')
```

```python
dfSeed2 = dfSeed.drop(['Seed'], axis = 1)
```

```python
style.use('fivethirtyeight')

graph = dfSeed.plot(kind='scatter', x = 'Seed',y = 'Correlation', s = 25, figsize = (12, 8))

# Setting rules for labeling axes
graph.tick_params(axis = 'both', which = 'major', labelsize = 12)

# Title

# Y value is continuous to values on Y-axis. For instance, 
# if the maximum value is 60, the text can begin at 61.

# Alpha indicates text opacity

graph.text(x = 0, y = 0.74, s = "Seed Results",
               fontsize = 24, weight = 'bold', alpha = .7)

# Creating different lines accross the graph (can be used for important slices of results)
graph.axhline(y = 0.35, color = 'black', linewidth = 1.25, alpha = .7)

# Subtitle
graph.text(x = 0, y = 0.69,
               s = "All values except for one are clustered around low values.\nThat one high value represents the correct seed (500).",
              fontsize = 14, alpha = .6)
```
