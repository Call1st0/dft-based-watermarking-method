```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# Importing CSV files - different seeds
impactFactorMarked = pd.read_csv('impactFactorMarked.csv')
impactFactorMarked = impactFactorMarked.drop(['Unnamed: 0'], axis = 1)

impactFactorGCR = pd.read_csv('impactFactorGCR.csv')
impactFactorGCR = impactFactorGCR.drop(['Unnamed: 0'], axis = 1)
```

```python
style.use('fivethirtyeight')

graph = impactFactorMarked.boxplot(grid = False, figsize = (12, 8))

# Setting rules for labeling axes
graph.tick_params(axis = 'both', which = 'major', labelsize = 12)

# Title

# Y value is continuous to values on Y-axis. For instance, 
# if the maximum value is 60, the text can begin at 61.

# Alpha indicates text opacity

graph.text(x = 1, y = 1.2, s = "Impact Factor Results - Marked Images",
               fontsize = 24, weight = 'bold', alpha = .7)

# Subtitle
graph.text(x = 1, y = 1.1,
               s = "...\n...",
              fontsize = 14, alpha = .6)
```

```python
style.use('fivethirtyeight')

graph = impactFactorGCR.boxplot(grid = False, figsize = (12, 8))

# Setting rules for labeling axes
graph.tick_params(axis = 'both', which = 'major', labelsize = 12)

# Title

# Y value is continuous to values on Y-axis. For instance, 
# if the maximum value is 60, the text can begin at 61.

# Alpha indicates text opacity

graph.text(x = 1, y = 1.2, s = "Impact Factor Results - GCR Images",
               fontsize = 24, weight = 'bold', alpha = .7)

# Subtitle
graph.text(x = 1, y = 1.1,
               s = "...\n...",
              fontsize = 14, alpha = .6)
```
