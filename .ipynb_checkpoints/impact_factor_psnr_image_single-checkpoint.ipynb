```python
from wmark import WaterMark
import skimage.measure as msr
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style
import pandas as pd
```

```python
watermark_object = WaterMark(5)
```

```python
img_original = watermark_object.imread('lena.tif')
```

```python
max_impact = 10001
steps = 500
range_max = int(max_impact/steps)+1

results = np.zeros([range_max, 3])
impact_factor = range(0, max_impact, steps)

for index in range(0, range_max):
    img_marked = watermark_object.embedMark(img_original, factor = impact_factor[index])
    psnr_value = msr.compare_psnr(img_original, img_marked)
    ssim_value = msr.compare_ssim(img_original, img_marked, multichannel = True)
    
    results[index, 2] = ssim_value
    results[index, 1] = psnr_value    
    results[index, 0] = impact_factor[index]
    
    print(results[index, 0], results[index, 1], results[index, 2])
```

```python
# https://www.ritchieng.com/creating-dataframe-from-objects/
pandas_array = ['Impact Factor', 'PSNR', 'SSIM']
pandas_dataframe = pd.DataFrame(results, columns=pandas_array)
```

```python
pandas_dataframe
```

```python
# https://www.dataquest.io/blog/making-538-plots/

style.use('fivethirtyeight')

fte_graph_1 = pandas_dataframe.plot(x = 'Impact Factor', y = 'PSNR', figsize = (12, 8))
fte_graph_1.tick_params(axis = 'both', which = 'major', labelsize = 16)

# Creating different lines accross the graph (can be used for important slices)
fte_graph_1.axhline(y = 40, color = 'black', linewidth = 1.1, alpha = .7)
fte_graph_1.axhline(y = 45, color = 'black', linewidth = 1.1, alpha = .7)

# Setting visibility of X-label
fte_graph_1.xaxis.label.set_visible(True)

# Title
fte_graph_1.text(x = 0, y = 61.8, s = "How Impact Factor affects PSNR Values",
               fontsize = 26, weight = 'bold', alpha = .7)

# Subtitle
fte_graph_1.text(x = 0, y = 56.5,
               s = "There is a clear correlation between the Impact Factor and\nPSNR Values, and it is quite consistent throughout the graph",
              fontsize = 18, alpha = .6)
```

```python
fte_graph_2 = pandas_dataframe.plot(x = 'Impact Factor', y = 'SSIM', figsize = (12, 8))
fte_graph_2.tick_params(axis = 'both', which = 'major', labelsize = 16)

# Creating different lines accross the graph (can be used for important slices)
# fte_graph_2.axhline(y = 0.5, color = 'black', linewidth = 1.1, alpha = .7)
# fte_graph_2.axhline(y = 1, color = 'black', linewidth = 1.1, alpha = .7)

# Setting visibility of X-label
fte_graph_2.xaxis.label.set_visible(True)

# Title
fte_graph_2.text(x = 0, y = 1.06, s = "How Impact Factor affects SSIM Values",
               fontsize = 26, weight = 'bold', alpha = .7)

# Subtitle
fte_graph_2.text(x = 0, y = 1.03,
               s = "There is a clear correlation between the Impact Factor and\nSSIM Values, and it is quite consistent throughout the graph",
              fontsize = 18, alpha = .6)
```

```python
# Colors: https://github.com/matplotlib/matplotlib/blob/38be7aeaaac3691560aeadafe46722dda427ef47/lib/matplotlib/mpl-data/stylelib/fivethirtyeight.mplstyle

plt.style.use('fivethirtyeight')

fig, ax1 = plt.subplots(figsize=(12,8))

ax1.set_xlabel('Impact Factor')
ax1.set_ylabel('PSNR', color = '#008fd5', zorder = 3)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('SSIM', color = '#fc4f30', zorder = 4)

ax1.grid(True)
ax2.grid(True)

ax1.plot(pandas_dataframe['Impact Factor'], pandas_dataframe['PSNR'], color = '#008fd5')
ax2.plot(pandas_dataframe['Impact Factor'], pandas_dataframe['SSIM'], color = '#fc4f30')
```
